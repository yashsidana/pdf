```python
import kagglehub
import pandas as pd
import numpy as np

# ------------------------------------------------------------
# Load dataset
# ------------------------------------------------------------
path = kagglehub.dataset_download("shrutibhargava94/india-air-quality-data")
df = pd.read_csv(f"{path}/data.csv", encoding="ISO-8859-1", low_memory=False)

df.columns = df.columns.str.strip()
no2_col = [c for c in df.columns if "NO2" in c.upper()][0]
x = df[no2_col].dropna().astype(float).values


# ------------------------------------------------------------
# FUNCTION: compute parameters for a given roll number
# ------------------------------------------------------------
def learn_parameters(x, r):
    ar = 0.05 * (r % 7)
    br = 0.3 * ((r % 5) + 1)

    # Correct transformation (as per assignment)
    z = x + ar * np.sin(br * x)

    # Learn PDF parameters
    mu = np.mean(z)
    sigma2 = np.var(z)
    lambda_param = 1 / (2 * sigma2)
    c = np.sqrt(lambda_param / np.pi)

    return mu, lambda_param, c



r1 = 102303973


mu1, lam1, c1 = learn_parameters(x, r1)

print("\nRoll Number:", r1)
print("mu (μ)     =", mu1)
print("lambda (λ) =", lam1)
print("c          =", c1)


```
